```python
import pandas as pd
import re
from collections import Counter 
```

```python
import json
file = open("./fin_texts_eng-poetry.json", "r")
j = json.load(file)
lst = []
for i in j:
    lst.append([i["text_ru"], i["text_en"], i["url"]])
```

```python
df = pd.DataFrame(lst, columns=["text_ru","text_en", "url"])
```

```python
len(df.url.unique())
```

```python
df.text_en = df.text_en.apply(lambda x: re.sub("(<i>)|(<b>)|(</i>)|(</b>)", "", x))
df.text_ru = df.text_ru.apply(lambda x: re.sub("(<i>)|(<b>)|(</i>)|(</b>)", "", x))

df.text_en = df.text_en.apply(lambda x: re.sub("<.*?>", "", re.sub("<.*?>.*?</.*?>", "", x)))
df.text_ru = df.text_ru.apply(lambda x: re.sub("<.*?>", "", re.sub("<.*?>.*?</.*?>", "", x)))

df.text_ru = df.text_ru.apply(lambda x: re.sub(r"\r\n", "\n", x))
df.text_ru = df.text_ru.apply(lambda x: re.sub(r"\n+", "\n", x))
df.text_ru = df.text_ru.apply(lambda x: re.sub(r"^\n", "", x))
df.text_ru = df.text_ru.apply(lambda x: re.sub(r"\n$", "", x))
df.text_ru = df.text_ru.apply(lambda x: re.sub(r"\n\s+\n", "\n", x))
df.text_ru = df.text_ru.apply(lambda x: re.sub(r"<pre>|</pre>", "", x))

df.text_en = df.text_en.apply(lambda x: re.sub(r"\r\n", "\n", x))
df.text_en = df.text_en.apply(lambda x: re.sub(r"\n+", "\n", x))
df.text_en = df.text_en.apply(lambda x: re.sub(r"^\n", "", x))
df.text_en = df.text_en.apply(lambda x: re.sub(r"\n$", "", x))
df.text_en = df.text_en.apply(lambda x: re.sub(r"\n\s+\n", "\n", x))
df.text_en = df.text_en.apply(lambda x: re.sub(r"<pre>|</pre>", "", x))
```

```python
df = df[df.text_en.str.len() > 10]
df = df[df.text_ru.str.len() > 10]
df = df.reset_index(drop=True)
```

```python
с = 0
lst = []
for idx, (ru, en, url) in df.iterrows():
    if "перевод" in ru.split("\n")[-1].lower():
        norm_text = "\n".join(ru.split("\n")[:-1])
        df.iloc[idx].text_ru = norm_text
    if "перевод" in ru.split("\n")[-2].lower():
        norm_text = "\n".join(ru.split("\n")[:-2])
        df.iloc[idx].text_ru = norm_text
    if "пер." in ru.split("\n")[-1].lower():
        norm_text = "\n".join(ru.split("\n")[:-1])
        df.iloc[idx].text_ru = norm_text
    if "пер." in ru.split("\n")[-2].lower():
        norm_text = "\n".join(ru.split("\n")[:-2])
        df.iloc[idx].text_ru = norm_text
```

```python
sep = "\n#new_poem#\n"
translation_sep = "\n#translation#\n"
url_sep = "\n#url#\n"
```

```python
lst = []
with open("eng-poetry-new_bad.txt", "r") as file:
    text = file.read()
    for pack in list(filter(None, text.split(sep))):
        url = pack.split(url_sep)[1]
        other = pack.split(url_sep)[0]
        en = other.split(translation_sep)[-1]
        for translation in other.split(translation_sep)[:-1]:
            lst.append([translation, en, url])
```

```python
df.text_ru = df.text_ru.apply(lambda x: re.sub("<.*?>.*?</.*?>" , "", x))
df.text_ru = df.text_ru.apply(lambda x: re.sub("<.*?>" , "", x))
df.text_ru = df.text_ru.apply(lambda x: re.sub("\r\n" , "\n", x))
df.text_ru = df.text_ru.apply(lambda x: re.sub("\n+" , "\n", x))

df.text_en = df.text_en.apply(lambda x: re.sub("<.*?>.*?</.*?>" , "", x))
df.text_en = df.text_en.apply(lambda x: re.sub("<.*?>" , "", x))
df.text_en = df.text_en.apply(lambda x: re.sub("\r\n" , "\n", x))
df.text_en = df.text_en.apply(lambda x: re.sub("\n+" , "\n", x))

for i in sorted(set_sep, key=lambda x: len(x), reverse=True):
        df.text_ru = df.text_ru.apply(lambda x: x.replace("Перевод" + i, "ПЕРЕВОД"))
        df.text_ru = df.text_ru.apply(lambda x: x.replace("перевод" + i, "ПЕРЕВОД"))
        df.text_ru = df.text_ru.apply(lambda x: x.replace("пер." + i, "ПЕРЕВОД"))
```
