```python
import pandas as pd
import json
import re
```

```python
import json
file = open("./texts_soleil_v1.json", "r")
j = json.load(file)
lst = []
for i in j:
    lst.append([i["article"], i["original"], i["url"]])
```

```python
df = pd.DataFrame(lst, columns=["article","original","url"])
```

```python
def what_lang(string):
    en_rate = len(re.findall(r"[a-zA-Z ]", string)) / len(string)
    ru_rate = len(re.findall(r"[а-яА-Я ]", string)) / len(string)
    if en_rate > ru_rate:
        return "en"
    else:
        return "ru"
```

```python
lst_en_ru = []
for i in range(len(df)):
    if (what_lang(df.iloc[i].article) == "en" and what_lang(df.iloc[i].original) == "en") or \
            (what_lang(df.iloc[i].article) == "ru" and what_lang(df.iloc[i].original) == "ru"):
        pass
    elif what_lang(df.iloc[i].article) == "en":
        lst_en_ru.append([df.iloc[i].article, df.iloc[i].original, df.iloc[i].url])
    elif what_lang(df.iloc[i].original) == "en":
        lst_en_ru.append([df.iloc[i].original, df.iloc[i].article, df.iloc[i].url])
    else:
        pass
```

```python
data = pd.DataFrame(lst_en_ru, columns=["en","ru","url"])
```

```python
data["en-ru"] = data.en.str.split("\n").str.len() - data.ru.str.split("\n").str.len()
```

```python
sep = "\n\n\n#new_poem#\n\n\n"
translation_sep = "\n#translation#\n"
url_sep = "\n#url#\n"
diff_sep = "\n#diff#\n"

with open("tania_lookup.txt", "w") as file:
    for idx, (ru, en, url, diff) in data.iterrows():
        ru = re.sub("\n+", "\n", ru)
        ru = re.sub("\n+$", "", ru)
        ru = re.sub("^\n+", "", ru)
        en = re.sub("\n+", "\n", en)
        en = re.sub("\n+$", "", en)
        en = re.sub("^\n+", "", en)
        
        file.write(sep)
        file.write(diff_sep)
        file.write(str(diff))
        file.write(ru)
        file.write(translation_sep)
        file.write(en)
        file.write(url_sep)
        file.write(url)
        file.write(diff_sep)
        file.write(str(diff))
```
